```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

Papers = pd.read_csv("~/datasets/s4/MAG/Papers.csv")
```

##### Can you create a line plot with matplotlib or seaborn showing trend of the number of papers between 2000 and 2010 in MAG papers dataset? 

```python
ax = plt.axes()
ax.ticklabel_format(useOffset=False)

Papers_10years = Papers[(Papers['Year'] > 2000) & (Papers['Year'] <= 2010)]
plt.plot(Papers_10years.groupby('Year')['PaperId'].count());
plt.xlabel('Year');
plt.ylabel('Number of Papers');
plt.title('Year vs. Number of Papers');
```

### Advanced (optional)

##### Using matplotlib or seaborn, can you show the trend of citation numbers between 1950 and 2020 for different types of documents?

```python
ax = plt.axes()
ax.ticklabel_format(useOffset=False)

Papers_70years = Papers[(Papers['Year'] > 1950) & (Papers['Year'] <= 2020)][['DocType', 'Year', 'CitationCount']].\
        groupby(['DocType', 'Year']).count().reset_index()

sns.lineplot(data=Papers_70years, x='Year', y='CitationCount', hue='DocType');
plt.legend(loc='upper left');
```
